# 1. **인공지능의 언어**

##  * 언어번역 서비스

인공지능의 언어 분야에 관련된 서비스중에서 가장 눈에 띄는 것은 **번역서비스**라고 생각한다. 

언어는 소통에 있어서 가장 중요한 것인데 서로 다른 언어를 사용하는 사람들끼리는 소통이 매우 힘들어지고 

그런 사람과 소통을 하기 위해서는 그 사람의 언어를 배워야 하는데 언어를 배우기 위해서는 오랜 시간과 노력이 필요하기 때문에 한사람이 배울 수 있는 언어는 아무래도 한계가 있기 마련이다.

그러나 언어번역 서비스는 그러한 한계를 언어가 학습된 인공지능이 해결해주어서 우리가 언어가 다른 지역을 가더라도 그 지역 사람과 소통이 가능하게끔 하여 주어서 우리가 느끼는 **가장 큰 불편함**을 해결해 줄 수 있기때문이다.

현재 대중적으로 많이 알려져있는 번역서비스인 구글 번역과 네이버 파파고를 사용해보면 과거 번역기를 사용했을 때보다 훨씬 더 자연스럽고 매끄럽게 해당 언어를 표현해주어서 실제 번역한 것과 크게 다르지 않은 모습을 보여준다.

그러나 일상생활에 사용되는 문장들에 비해서 소설 등에 사용되는 표현들은 번역하였을 때 그 표현이 주는 느낌을 다른 언어로 전달하는데 많이 부족하여서 번역기로 번역한 표현을 읽었을 때 그 표현이 와 닿지 안게 번역되는 경우가 많아 아쉬운 점도 있었다.


그렇기 때문에 그러한 표현까지도 해당 언어로 표현할 수 있게 되어 서로 다른 언어를 사용하는 사람들의 문학까지 더욱 자유롭게 볼 수 있게 만들어 보고 싶어졌다.






 # 2. **인공지능의 음성**

 ## * 음성 인식 비서 서비스

 인공지능의 음성 분야에 관련된 서비스 중 우리 주변에서 가장 많이 볼 수 있고 또  우리가 인공지능을 생각했을 때 가장 먼저 떠오르는 것이 바로 이 비서 서비스라고 생각한다.

듣고 말하는 것은 소통의 기본이기에 우리가 인공지능에게 말을 하고 인공지능이 그 말을 듣고 답을 주는 것이 인공지능과 소통을 한다는 느낌을 가장 강하게 느낄 수 있기 때문이다.

현재 가장 많이 사용되는 음성 서비스인 스마트폰 비서 서비스 Siri와 빅스비를 직접 사용하다 보면 내가 마치 살아있는 대상을 상대하는 기분을 느낄 수 있는데 그럴 때면 지금 누군가와 같이 있다는 느낌을 얻을 수 있었다.

요즘 사람들은 혼자 사는 인구가 많아져서 외로움을 느끼는 경우가 많은데 이러한 서비스가 더욱 발전한다면 더 이상 혼자 살아도 외로움을 그 전보다 덜 느낄 수 있게 될 거라고 생각된다.

그러나 아직 이 비서 서비스는 인공지능이라 말하기엔 너무 딱딱하고 늘 정해진 답만 내놓기 때문에 그러한 느낌을 받기 위해서는 아직도 많은 발전이 필요한데 나는 이러한 서비스를 발전시켜서 나만의 비서를 만들어 보고 싶다.




 


# 3. **인공지능의 이미지**

## * 이미지 인식 기술 

인공지능의 이미지와 관련된 기술 중에서 눈에 띄는 기술은 바로 이미지 인식 기술이다.

이미지 인식 기술을 해당 이미지를 인식하여 그 이미지에 찍혀있는 대상의 정보를 파악해서 알려주는 기술이다.

요즘같이 새로운 물건이 계속해서 만들어지는 시대에 살다 보면 온갖 처음 보는 물건들이 많은데 이 기술을 사용하면 그 물건의 사진을 찍어 검색하여 그 물건의 이름과 용도들을 알 수 있게끔 해주는 게 매우 편리한 기능이라고 생각한다.

물론 아직 실생활에 도움이 될 정도로 이 기술이 발전하지 않아서 많이 사용되지 않고 있지만 이 기술이 발전하고 인공지능이 더욱더 많은 학습을 거친다면 우리 생활에 많은 도움이 될 거라고 생각된다.



# 4. **인공지능의 자율주행**

## * 자율주행 자동차

인공지능 분야에서 가장 논란도 많고 기대도 많은 게 바로 이 자율주행 분야라고 생각한다.

위에 적힌 다른 분야 서비스들과 다르게 자율주행은 사용자의 안전과 관련이 있고 또 자동차는 우리가 가장 많이 사용하는 이동수단이기 때문에 더욱더 많은 관심을 받고 있다.

이 기술 오래전부터 연구해왔고 요즘 들어 자율주행 자동차가 상용화되고 있는데 최근 세종시에서 자율주행 버스를 운행하여 많은 관심을 받고 있다.

개인적으로 이 기술이 빨리 상용화되었으면 좋겠다. 그 이유는 우리나라는 운전면허 취득이 쉽고 안전에 관한 주의가 많이 부족해서 자동차 관련 사고가 잦은데 이 기술이 상용화되면 그러한 안전사고가 많이 줄어들고 또 교통의 효율이 증가하여 이동 시간이 단축될 수 있기 때문이다.

그러나 우리들의 안전과 관련되어 있는 만큼 더욱더 조심스럽게 다가가야 한다고 생각한다.
